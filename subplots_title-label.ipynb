```python
import matplotlib.pyplot as plt
import numpy as np

x = np.linspace(0,5,11)
y = x**2
```

```python
fig, axes = plt.subplots()
axes.plot(x,y)
plt.show()
```

```python
fig, axes = plt.subplots(nrows=1, ncols=2, figsize=(7,5))
axes[0].set_title('Relation between X - Y')
axes[0].set_xlabel('X')
axes[0].set_ylabel('Y')
axes[0].plot(x,y,'b', label='$x**2$')
axes[0].legend()
axes[1].set_title('Relation between Y - X')
axes[1].set_xlabel('Y')
axes[1].set_ylabel('X')
axes[1].plot(y,x,'r')
plt.show()
```

```python
fig, (axes1,axes2) = plt.subplots(nrows=1, ncols=2)
axes1.plot(x,y,'b')
axes2.plot(y,x,'r')
plt.show()
```

```python
fig, axes = plt.subplots(2,4)

#genera un trazo accediendo a las graficas a traves del indice de la matriz

axes[0,0].plot(x,y)
axes[0,1].plot(y,x, 'r')
axes[0,2].plot(x, np.cos(x), 'y')
axes[0,3].plot(y, np.sin(y), 'b')

fig.tight_layout() #mejora la visualización de los ejes de cada gráfico
```
