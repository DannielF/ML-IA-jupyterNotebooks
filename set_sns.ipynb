```python
import seaborn as sns
import matplotlib.pyplot as plt
```

## Set is a Seaborn method to allow you set styles, font and graphics color

```python
sns.barplot(x=['A','b','C'], y=[1,3,2])
plt.show()
```

```python
sns.set(style='dark', palette='dark', font="Serif", font_scale=1)
sns.barplot(x=['A','b','C'], y=[1,3,2])
plt.show()
```
